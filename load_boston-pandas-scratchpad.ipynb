```python
labels = np.array(fopen[13])
features = np.array(fopen.drop([13, 0], 1))
features = np.column_stack((np.ones(506), features))
weights = .01*np.ones(features.shape[1])
for i in xrange(10000):
    g = np.dot(labels-np.dot(features, weights), features)/features.shape[0]
    mat = np.dot(features, weights)
    error = labels - mat
    grad = np.dot((error),(features))
    gradn = grad/features.shape[0]
    weights += .0001*gradn
```
