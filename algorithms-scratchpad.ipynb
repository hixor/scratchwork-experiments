```python
# insertion sort

def insertionSort(lst):
    new = list()
    new.append(lst[0])
    for k in range(1,len(lst)):
        i = 0
        while i <= len(new)-1:
            #print i
            #print lst[k]
            if lst[k] > new[i]:
                i = i+1
            else:
                new.append(None)
                j = len(new)-1
                while j >= i+1:
                    new[j] = new[j-1]
                    j = j-1
                new[i] = lst[k]
                break
        if i == len(new):
            new.append(lst[k])
    return new
```

```python
import random

for j in range(10):
    lst = list()
    for i in range(15):
        lst.append(random.randint(-100,100))
    print "Unsorted: ", lst
    print "Sorted: ", insertionSort(lst)
```

```python
#mergesort

def mergeSort(lst):
    if len(lst) == 1:
        return lst
    else:
        left = mergeSort(lst[:len(lst)/2])
        right = mergeSort(lst[len(lst)/2:])
    
        i = 0
        j = 0
        k = 0
        merged = list()
        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                merged.append(left[i])
                i = i+1
            else:
                merged.append(right[j])
                j = j+1
        while i < len(left):
            merged.append(left[i])
            i = i+1
        while j < len(right):
            merged.append(right[j])
            j = j+1
        return merged
        
    return merged
```

```python
# maximum Subarray

def maxSub(array):
    if len(array) == 1:
        return array
    else:
        left = maxSub(array[:len(array)/2])
        right = maxSub(array[len(array)/2:])
        #mid = maxMidSub(array)
        
        if sum(left) > sum(right):
            return left
        else:
            return right
```

```python
# maximum subarray, textbook version

def maxCenter(array,i,j,mid):
    maxleft = -float('inf')
    maxright = -float('inf')
    maxi = i
    maxj = j
    for k in range(i,mid,-1):
        if array[k]+array[k-1] > maxleft or maxleft == None:
            maxleft = array[k]+array[k-1]
            maxi = k-1
    for l in range(mid,j):
        if array[l]+array[l+1] > maxright or maxright == None:
            maxright = array[l]+array[l+1]
            maxj = l+1
    return (maxi,maxj,maxleft+maxright)

def maxSub1(array,i,j):
    if i == j:
        return (array[i],i,j)
    else:
        left = maxSub1(array,i,j/2)
        right = maxSub1(array,i/2+1,j)
        center = maxCenter(array,i,j,i/2)
        print left
        print right
        print center
        
        if left[0] >= right[0] and left[0] >= center[2]:
            return (left[1],left[2],sum(left[(left[1]):left[2]+1]))
        elif right[0] >= left[0] and right[0] >= center[2]:
            return (right[1],right[2],sum(right[(right[1]):right[2]+1]))
        else:
            return (center[0],center[1],center[2])
```
